```python

# correct: sum the boolean Series, not the string
churn_count = (df["Churn"] == "yes").sum()
# or, case-insensitive and trimmed:
churn_count = df["Churn"].str.strip().str.lower().eq("yes").sum()
print(churn_count)
```

```python
# Normalize column names/values and compute churn rate
df.columns = df.columns.str.strip()                      # remove extra spaces from column names

# Find churn column robustly
if 'Churn' in df.columns:
    churn_col = 'Churn'
elif 'churn' in df.columns:
    churn_col = 'churn'
else:
    # fallback: lowercase all columns then try
    df.columns = df.columns.str.lower()
    churn_col = 'churn' if 'churn' in df.columns else None
    if churn_col is None:
        raise KeyError("No churn column found in dataframe")

# Normalize values and compute
churn_bool = df[churn_col].astype(str).str.strip().str.lower().eq('yes')
churn_rate = churn_bool.mean() * 100.0

# Print precise percent
print(f"Overall Churn Rate: {churn_rate:.2f}%")

# Print integer-range like "26-27%" (use floor/ceil)
import math
low, high = math.floor(churn_rate), math.ceil(churn_rate)
print(f"Overall Churn Rate (range): {low}-{high}%")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.boxplot(x="Churn", y="tenure", data=df)

plt.title("Tenure vs Churn")
plt.xlabel("Churn")
plt.ylabel("Tenure (months)")

plt.tight_layout()
plt.show()
```

```python
# use the existing variable that holds the churn column name (handles case sensitivity)
df.groupby(churn_col)["tenure"].mean().round(1)
```

Customers with shorter tenure are significantly more likely to churn.
On average, churned customers have been with the company for 18 months, compared to 38 months for retained customers.
This indicates that the highest churn risk occurs in the early stages of the customer lifecycle, highlighting the importance of strong onboarding and early engagement strategies.

```python
plt.figure(figsize=(7,4))
sns.countplot(x="Contract", hue="Churn", data=df)

plt.title("Churn by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Number of Customers")
plt.xticks(rotation=15)

plt.tight_layout()
plt.show()
```

Contract type is one of the strongest predictors of churn.
Customers on month-to-month contracts exhibit significantly higher churn rates compared to those on 1-year or 2-year contracts, indicating lower commitment and higher sensitivity to price or service changes.
Longer-term contracts appear to be an effective retention mechanism.

```python
(
    df.groupby("Contract")["Churn"]
      .apply(lambda x: (x == "Yes").mean() * 100)
      .round(2)
      .sort_values(ascending=False)
)
plt.figure(figsize=(7,4))
sns.countplot(x="PaymentMethod", hue="Churn", data=df)
plt.title("Churn by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Count")
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x="InternetService", hue="Churn", data=df)

plt.title("Churn by Internet Service Type")
plt.xlabel("Internet Service")
plt.ylabel("Number of Customers")

plt.tight_layout()
plt.show()
```

Churn varies significantly by internet service type.
